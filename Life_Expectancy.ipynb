```python
data.Alcohol = data.Alcohol.fillna(data.Alcohol.mean())
data.Hepatitis_B = data.Hepatitis_B.fillna(data.Hepatitis_B.mean())
data.BMI = data.BMI.fillna(data.BMI.mean())
data.Polio = data.Polio.fillna(data.Polio.mean())
data.Total_expenditure = data.Total_expenditure.fillna(data.Total_expenditure.mean())
data.Diphtheria = data.Diphtheria.fillna(data.Diphtheria.mean())
data.HIV_AIDS = data.HIV_AIDS.fillna(data.HIV_AIDS.mean())
data.GDP = data.GDP.fillna(data.GDP.mean())
data.Population = data.Population.fillna(data.Population.mean())
data.Income_composition_of_resources = data.Income_composition_of_resources.fillna(data.Income_composition_of_resources.mean())
data.Schooling = data.Schooling.fillna(data.Schooling.mean())
data.Adult_Mortality = data.Adult_Mortality.fillna(data.Adult_Mortality.mean())
```

Encoding using Label Encoder 

```python
from sklearn.preprocessing import LabelEncoder 
encoder = LabelEncoder()
```

```python
data['Country_Encoded'] = encoder.fit_transform(data['Country'])
data['Status_Encoded'] = encoder.fit_transform(data['Status'])
```

Dropping the columns that are objects and leaving the encoded versions

```python
data = data.drop(['Country','Status', 'infant_deaths', 'Adult_Mortality', 'GDP', 'Alcohol', 'percentage_expenditure'], axis = 1)
```

Training the data using different models

```python
x = data.drop("Life_expectancy", axis=1)
y = data['Life_expectancy']
```

```python
from sklearn.feature_selection import SelectKBest, f_regression, RFE

select_k_best = SelectKBest(score_func=f_regression, k=10)  # select top 10 features
x_train_k_best = select_k_best.fit_transform(x_train, y_train)
x_test_k_best = select_k_best.transform(x_test)

# Fit and evaluate the model with selected features
lr.fit(x_train_k_best, y_train)
y_test_pred = lr.predict(x_test_k_best)
mse_k_best = mean_squared_error(y_test, y_test_pred)
print(f'MSE with SelectKBest: {mse_k_best}')

# Recursive Feature Elimination (RFE)
rfe = RFE(estimator=LinearRegression(), n_features_to_select=10)
x_train_rfe = rfe.fit_transform(x_train, y_train)
x_test_rfe = rfe.transform(x_test)

# Fit and evaluate the model with RFE selected features
lr.fit(x_train_rfe, y_train)
y_test_pred = lr.predict(x_test_rfe)
mse_rfe = mean_squared_error(y_test, y_test_pred)
print(f'MSE with RFE: {mse_rfe}')
```
